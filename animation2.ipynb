```python
# アニメーションを作ろう
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML
```

```python
fig = plt.figure()
img_buffer =[]

x = 0
y = 0
plt.xlim([0,10])
plt.ylim([-5,5])

for t in range(30):
    x = x+1
    if x > 8:
        x =8
    y = y
    img = plt.plot(x, y, linewidth=0, marker ='o', markersize=15, color ='r')
    img_buffer.append(img)
    
ts = 100 
ani = animation.ArtistAnimation(fig, img_buffer, interval=ts, repeat=False)

plt.close()
HTML(ani.to_jshtml())

    
```

```python
fig = plt.figure()
img_buffer =[]

x = 0
y = 0
plt.xlim([0,10])
plt.ylim([-5,5])

for t in range(30):
    x = x+1
    if x > 8:
        x = x-1
    if x < 0:
        x= x+1
    y = y
    img = plt.plot(x, y, linewidth=0, marker ='o', markersize=15, color ='r')
    img_buffer.append(img)
    
ts = 100 
ani = animation.ArtistAnimation(fig, img_buffer, interval=ts, repeat=False)

plt.close()
HTML(ani.to_jshtml())

    
```
