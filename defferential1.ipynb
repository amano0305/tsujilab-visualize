```python
# グラフを書いてみよう
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
def ode(t, y, gamma):
    dx =gamma*y[0]
    return [dx]

gamma = 1.0
Tend = 10.0
x0 = 10

sol = solve_ivp(fun=ode, t_span=[0, Tend], y0=[x0], args=[gamma], dense_output=True)
```

```python
tc =np.linspace(0, Tend, 100)
xc = sol.sol(tc)

plt.plot(tc,xc[0], label="x(t): population")
plt.xlabel("time")
plt.ylabel("number")
plt.legend()
plt.grid()
```
