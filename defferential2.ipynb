```python
# グラフを書いてみよう
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
def ode(t, y, a, b, c, d):
    dx =a*y[0]-b*y[0]*y[1]
    dy =-c*y[1] + d*y[0]*y[1]
    return [dx, dy]

a, b, c, d = 2, 1, 3, 1
Tend =10.0
x0, y0 = 5, 1

sol =solve_ivp(fun=ode, t_span=[0, Tend], y0=[x0, y0], args=[a, b, c, d], dense_output=True)

```

```python

tc =np.linspace(0, Tend, 100)
yc =sol.sol(tc)

plt.plot(tc,yc[0], label="x(t): Prede")
plt.plot(tc,yc[1], label="y(t): Predatori")

plt.xlabel("time")
plt.ylabel("nuber")
plt.legend()
plt.grid()
```
