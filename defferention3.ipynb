```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
from scipy.integrate import solve_ivp

def odeSIR(t, y, bete, gamma):
    dS = -beta*y[0]*y[1]
    dI = beta*y[0]*y[1]-gamma*y[1]
    dR = gamma*y[1]
    return [dS,dI, dR]

beta,gamma = 0.01, 0.5
Tend =3.0
coef= 10
S0 = coef*gamma/beta
I0 = 10
R0 = 0


sol =solve_ivp(fun=odeSIR,t_span=[0, Tend], y0=[S0, I0, R0], args=[beta, gamma], dense_output=True)
```

```python
tc =np.linspace(0, Tend, 100)
yc =sol.sol(tc)

plt.plot(tc,yc[0], label="S(t): susceptibles")
plt.plot(tc,yc[1], label="I(t): infectives")
plt.plot(tc,yc[2], label="R(t): recovered")

plt.xlabel("time")
plt.ylabel("nuber")
plt.legend()
plt.grid()
```
