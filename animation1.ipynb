```python
# アニメーションを作ろう
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML
```

```python
fig = plt.figure()
img_buffer =[]

for t in range(30):
    x = t 
    y = x 
    img =plt.plot(x, y, linewidth=0, marker='o',markersize=15, color='r')
    img_buffer.append(img)
 
ts = 100 #animationの間隔
ani = animation.ArtistAnimation(fig, img_buffer, interval=ts, repeat=False)

plt.close()
HTML(ani.to_jshtml())
    
```
